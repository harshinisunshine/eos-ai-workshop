**SMILEY AND HEART CODE-NUMPY AND PYTHON**

```python
#smiley code
import numpy as np
import matplotlib.pyplot as plt
mat1=np.fliplr(np.eye(4))
mat2=np.zeros((4,4))
mat2[0,:]=1
mat3=np.zeros((4,4))
mat3[0,:]=1
mat4=np.eye((4))
hs1=np.hstack((mat1,mat2,mat3,mat4))
mat5=np.zeros((4,4))
mat5[:,0]=1
mat6=np.zeros((4,4))
mat7=np.zeros((4,4))
mat8=np.zeros((4,4))
mat8[:,3]=1
hs2=np.hstack((mat5,mat6,mat7,mat8))
mat9=np.zeros((4,4))
mat9[:,0]=1
mat10=np.zeros((4,4))
mat11=np.zeros((4,4))
mat12=np.zeros((4,4))
mat12[:,3]=1
hs3=np.hstack((mat9,mat10,mat11,mat12))
mat13=np.eye((4))
mat14=np.zeros((4,4))
mat14[3,:]=1
mat15=np.zeros((4,4))
mat15[3,:]=1
mat16=np.fliplr(np.eye(4))
hs4=np.hstack((mat13,mat14,mat15,mat16))
vs=np.vstack((hs1,hs2,hs3,hs4))
vs[3:5,3:5]=1
vs[3:5,11:13]=1
vs[11,3]=1
vs[11,12]=1
vs[12,4:12]=1
print(vs)

plt.imshow(vs,cmap="binary")
plt.axis('off')
plt.title("original smiley")
plt.show()

noise=np.random.randn(16,16)*0.1
vs_n=vs+noise
vs_f=vs_n.reshape(-1)
vs_2=vs_f.reshape(16,16)
vs_2d=1-vs_2


vs_rgb = np.stack([vs_2d, vs_2d, vs_2d], axis=-1)
print(f"Original shape: {vs.shape}")
print(f"New RGB shape: {vs_rgb.shape}")
plt.imshow(vs_rgb)
plt.title("Noisy RGB Smiley")



```
